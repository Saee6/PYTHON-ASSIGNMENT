```python
A = 90
B = 3

count = 0

while A % B == 0:
    A = A // B
    count += 1

if count > 0:
    print("Purely divisible", count, "times")
else:
    print("Not purely divisible")

```

```python
numbers = list(range(1, 26))

for num in numbers:
    if num % 3 == 0:
        print(f"{num} → divisible by 3")
    else:
        print(f"{num} → not divisible by 3")

```
